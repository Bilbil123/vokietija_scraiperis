```python
import pandas as pd

# Nuskaitome failą "Bücher .csv"
df = pd.read_csv('Bücher .csv')

# Sukuriame naują stulpelį "Trumpas aprašymas", paimdami pirmus 85 simbolius iš "Aprašymas"
df['Trumpas aprašymas'] = df['Aprašymas'].str[:85]

# Išsaugome atnaujintą failą kaip "Bücher_atnaujintas.csv"
df.to_csv('Bücher_atnaujintas.csv', index=False, encoding='utf-8')

# Peržiūrime pirmas kelias eilutes, kad patikrintume
print(df.head())
```

```python
import pandas as pd

# Nuskaitome failą "Bücher_atnaujintas.csv"
df = pd.read_csv('Bücher_atnaujintas.csv')

# Pakeičiame "Telefonas nerastas" į "Telefon nicht gefunden" stulpelyje "Telefonas"
df['Telefonas'] = df['Telefonas'].replace('Telefonas nerastas', 'Telefon nicht gefunden')

# Išsaugome atnaujintą failą
df.to_csv('Bücher_atnaujintas.csv', index=False, encoding='utf-8')

# Peržiūrime pirmas kelias eilutes, kad patikrintume
print(df.head())
```

```python
import pandas as pd

# Nuskaitome failą "Bücher_atnaujintas.csv"
df = pd.read_csv('Bücher_atnaujintas.csv')

# Pakeitimai atitinkamuose stulpeliuose:
df['Svetainė'] = df['Svetainė'].replace('Svetainės adresas nerastas', 'Website-Adresse nicht gefunden')
df['El. paštas'] = df['El. paštas'].replace('El. paštas nerastas', 'E-Mail nicht gefunden')
df['Adresas'] = df['Adresas'].replace('Aprašymas nerastas', 'Adresse nicht gefunden')

# Išsaugome atnaujintą failą
df.to_csv('Bücher_atnaujintas.csv', index=False, encoding='utf-8')

# Peržiūrime pirmas kelias eilutes, kad patikrintume
print(df.head())
```
